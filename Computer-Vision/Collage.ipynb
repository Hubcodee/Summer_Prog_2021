```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading image file
img1 = cv2.imread('download.jpg')
```

```python
#resizing dimensions of image
img1 = cv2.resize(img1,(500,500))
```

```python
#reading image file
img2 = cv2.imread('photo.jpg')
```

```python
#resizing dimensions of image
img2 = cv2.resize(img2,(500,500))
```

```python
#using hstack() method to make horizontal collage
frame = np.hstack([img1,img2])
#using vstack() method to make vertical collage
frame2 = np.vstack([img1,img2])
```

```python
plt.imshow(frame)
```

```python
plt.imshow(frame2)
```
